```python
def convert_to_minutes(time_str):
    if pd.isna(time_str):
        return None  # Keep NaN if missing
    # Extend time_map to handle singular and plural forms, including days
    time_map = {"day": 1440, "days": 1440, "hr": 60, "hrs": 60, "min": 1, "mins": 1}
    
    # Split the time string into parts and calculate total minutes
    time_parts = time_str.split()
    return sum(int(num) * time_map[unit] for num, unit in zip(time_parts[:-1:2], time_parts[1::2]))

# Apply the function to the relevant columns
d['prep_time_mins'] = d['prep_time'].apply(convert_to_minutes)
d['cook_time_mins'] = d['cook_time'].apply(convert_to_minutes)
d['total_time_mins'] = d['total_time'].apply(convert_to_minutes)
```

```python
import numpy as np

def recommend_recipes(user_input, df, category_weight=2, keyword_boost_weight=0.1, top_n=5, filters=None, filter_weights=None):
    """
    Recommend recipes based on user input, applying optional filters and similarity scores.

    Args:
        user_input (str): Input string describing desired ingredients, categories, or keywords.
        df (pd.DataFrame): DataFrame containing recipe data with embeddings, categories, and nutritional information.
        category_weight (float): Weight to give category similarity compared to text similarity.
        keyword_boost_weight (float): Weight to boost similarity based on matching keywords in ingredients.
        top_n (int): Number of top recommendations to return.
        filters (dict, optional): Constraints for filtering recipes, with column names as keys and (min, max) as values.
                                  Example: {"calories": (300, 1000), "protein": (20, 40)}.
        filter_weights (dict, optional): Weights for the filters to prioritize specific columns during scoring.
                                         Default weight for all filters is 1.

    Returns:
        pd.DataFrame: DataFrame containing the top recommended recipes, with columns:
                      ['category', 'title', 'ingredients', 'calories', 'protein', 'fat',
                       'carbs', 'serving_info', 'total_time_mins', 'similarity',
                       'filter_match', 'final_score'].

    Nested Functions:
        boost_similarity: Calculate keyword-based boosts for similarity scores.
        calculate_filter_match: Calculate a normalized match score for each recipe based on filter constraints.
    """
    # Preprocess user input to emphasize category
    processed_input = f"{user_input} {user_input}"
    user_embedding = model.encode(processed_input)
    
    # Calculate cosine similarity for text and category
    recipe_embeddings = np.vstack(df['embedding'])
    text_similarities = cosine_similarity([user_embedding], recipe_embeddings).flatten()

    df['category_embedding'] = df['category'].apply(lambda x: model.encode(x))
    category_embeddings = np.vstack(df['category_embedding'])
    category_similarities = cosine_similarity([user_embedding], category_embeddings).flatten()

    # Combine similarities
    final_similarities = text_similarities + category_weight * category_similarities
    df['similarity'] = final_similarities

    # Fix ingredients if they are stored as strings
    if isinstance(df['ingredients'].iloc[0], str):
        df['ingredients'] = df['ingredients'].apply(lambda x: eval(x))

    # Define boost function for ingredient matches
    def boost_similarity(row, keywords):
        """
        Calculate a boost to similarity score based on keyword matches in ingredients.

        Args:
            row (pd.Series): A row of the DataFrame representing a recipe.
            keywords (list): List of keywords to match in the ingredient list.

        Returns:
            float: Boosted similarity score for the recipe.
        """
        ingredient_text = " ".join(row['ingredients']).lower()  # Join ingredients into a single string
        return sum(1 for keyword in keywords if keyword.lower() in ingredient_text) * keyword_boost_weight

    # Apply boost based on user keywords
    keywords = user_input.split()  # Extract keywords from user input
    df['boost'] = df.apply(lambda row: boost_similarity(row, keywords), axis=1)
    df['similarity'] += df['boost']  # Add boost to final similarity score

    # Calculate % match score for filters
    if filters:
        def calculate_filter_match(row, filters, filter_weights):
            """
            Calculate a normalized match score for a recipe based on its adherence to filter constraints.
            
            Ex: 
                For each filter column (e.g., protein), if a recipe's value is within the range, it contributes a 100% score for that column.
                If the value is outside the range, the score decreases linearly as the value moves further away.

            Args:
                row (pd.Series): A row of the DataFrame representing a recipe.
                filters (dict): Constraints with column names as keys and (min, max) ranges as values.
                filter_weights (dict): Weights for filter columns to prioritize specific attributes.

            Returns:
                float: Normalized filter match score for the recipe.
            """
            match_scores = []
            for column, (min_val, max_val) in filters.items():
                if column in row:
                    value = row[column]
                    range_spread = max_val - min_val
                    # Handle edge cases: no range spread
                    if range_spread == 0:
                        score = 1.0 if value == min_val else 0.0
                    else:
                        if min_val <= value <= max_val:
                            score = 1.0
                        else:
                            # Calculate linear drop-off for out-of-range values
                            closest_bound = min_val if value < min_val else max_val
                            score = max(0, 1 - abs(value - closest_bound) / range_spread)
                    
                    # Apply column weight
                    weight = filter_weights.get(column, 1) if filter_weights else 1
                    match_scores.append(score * weight)
            
            # Normalize by the sum of weights
            total_weight = sum(filter_weights.values()) if filter_weights else len(filters)
            return sum(match_scores) / total_weight

        # Compute % match score for each recipe
        filter_weights = filter_weights or {column: 1 for column in filters.keys()}  # Default weights = 1
        df['filter_match'] = df.apply(lambda row: calculate_filter_match(row, filters, filter_weights), axis=1)
    
    # Combine similarity and filter match into a final score
    df['final_score'] = df['similarity'] + df['filter_match']  # Combine scores (can be weighted)

    # Sort by final score and return top recipes
    recommended = df.sort_values(by='final_score', ascending=False).head(top_n)
    return recommended[['category', 'title', 'ingredients', 'calories', 'protein', 'fat', 'carbs', 'serving_info', 'total_time_mins', 'similarity', 'filter_match', 'final_score']]


```

```python
# User input
user_input = "Pork green onions bacon"

# Filters: high protein, low fat, short preparation time
filters = {
    "calories": (300, 1000),         # 300 to 1000 calories
    "protein": (20, 40),             # 20 to 40 grams of protein
    "fat": (0, 20),                  # Up to 20 grams of fat
    "total_time_mins": (0, 60)       # Up to 60 minutes of preparation time
}

# Weights for filters (optional, can prioritize some filters more than others)
filter_weights = {
    "calories": 1,
    "protein": 2,   # Prioritize protein
    "fat": 1,
    "total_time_mins": 1.5  # Slightly prioritize time
}

# Get recommendations
recommended_recipes = recommend_recipes(user_input, df, category_weight=2, keyword_boost_weight=0.1, top_n=5, filters=filters, filter_weights=filter_weights)

recommended_recipes
```
